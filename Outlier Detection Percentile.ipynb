```python
import pandas as pd
df = pd.read_csv('height.csv')
df.head()
```

```python
max_height = df['height'].quantile(0.95)
max_height
```

```python
df[df['height']>max_height]
```

```python
min_height = df['height'].quantile(0.05)
min_height
```

```python
df[df['height']<min_height]
```

```python
df[(df['height']< max_height) & (df['height']>min_height)]
```

### Getting data from Github.com

```python
url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/1_outliers/bhp.csv'
banglor = pd.read_csv(url, index_col = 0)
banglor
```

```python
max_price, min_price = banglor['price_per_sqft'].quantile([0.988, 0.00122])
```

```python
min_price, max_price
```

```python
banglor[banglor.price_per_sqft < min_price]
```

```python
with_out = banglor[(banglor['price_per_sqft']> min_price) & (banglor['price_per_sqft']< max_price)]
with_out.sample(5)
```
